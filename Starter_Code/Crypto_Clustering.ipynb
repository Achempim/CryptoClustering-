Libraries 

```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

Load CSV File

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv", 
    index_col="coin_id")
```

Display Summary Statistics 

```python
# Generate summary statistics
df_market_data.head(10)
```

Plot the data using hvplot

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

### Prepare the Data

Use the StandardScaler() module from scikit-learn to normalize the data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()
```

```python
# Fitt and transform on the original DataFrame's features
scaled_data = scaler.fit_transform(df_market_data)
```

Reset the index to match the "coin_id" from the original

```python
# Create a new DataFrame with scaled data, and set the "coin_id" as the index
df_scaled = pd.DataFrame(scaled_data, index=df_market_data.index, columns=df_market_data.columns)

# Show the first five rows of the scaled DataFrame
df_scaled.head(5)
```

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

Empty List

```python
# Create an empty list to store the inertia values
inertia_values = []
```

```python
# Create a for loop to compute the inertia for each k
for k in k_values:
    try:
        # 1. Create a KMeans model using the loop counter for the n_clusters
        kmeans = KMeans(n_clusters=k, random_state=0)
        
        # 2. Fit the model to the scaled data
        kmeans.fit(df_scaled)
        
        # 3. Append the model.inertia_ to the inertia list
        inertia_values.append(kmeans.inertia_)
    
    except Exception as e:
        # Log any errors encountered during the KMeans fitting process
        print(f"Error encountered with k={k}: {e}")
```

```python
# Now, ensure the lengths of k_values and inertia_values match
print(f"Length of k_values: {len(k_values)}")
print(f"Length of inertia_values: {len(inertia_values)}")
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    'k': k_values,
    'inertia': inertia_values
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Check the resulting DataFrame
print(df_elbow)
```

```python
# Plot a line chart with all the inertia values
df_elbow.hvplot.line(
    x='k', 
    y='inertia', 
    title="Elbow Curve to Find Optimal k", 
    xlabel="k (Number of Clusters)", 
    ylabel="Inertia"
)
```

```python
# Initialize the K-Means model using the best value for k (k=4, adjust based on elbow plot)
best_k = 4  # Replace with the value you find from the elbow plot
kmeans = KMeans(n_clusters=best_k, random_state=0)
```

```python
# Fit the K-Means model using the scaled DataFrame
kmeans.fit(df_scaled)
```

# Perform PCA and examine explained variance.

### Optimize Clusters with Principal Component Analysis.

#### Create a PCA Model Instance and Set n_components=3

```python
# Create a PCA model instance and set n_components=3
pca = PCA(n_components=3)
```

#### Use the PCA Model with fit_transform

```python
# Use the PCA model with fit_transform to reduce the scaled DataFrame down to three principal components
pca_data = pca.fit_transform(df_scaled)

# View the PCA data
pca_data[:5]  # Show the first 5 rows of the PCA data
```

#### Retrieve the Explained Variance

```python
# Retrieve the explained variance to determine how much information can be attributed to each principal component
explained_variance = pca.explained_variance_ratio_

# Print the explained variance for each principal component
print(f"Explained variance ratio: {explained_variance}")
```

```python
# Create an empty list to store inertia values
inertia_values_pca = []
```

```python
# Now, ensure the lengths of k_values and inertia_values match
print(f"Length of k_values: {len(k_values)}")
print(f"Length of inertia_values: {len(inertia_values_pca)}")
```

```python
# Initialize an empty list to store the inertia values
inertia_values_pca = []

# Create a for loop to compute the inertia for each k 
for k in k_values:
    try:
        # 1. Create a KMeans model using the loop counter for the n_clusters
        kmeans = KMeans(n_clusters=k, random_state=0)
        
        # 2. Fit the model to the scaled data
        kmeans.fit(df_pca)
        
        # 3. Append the model.inertia_ to the inertia_values_pca list
        inertia_values_pca.append(kmeans.inertia_)
    
    except Exception as e:
        # Log any errors encountered during the KMeans fitting process
        print(f"Error encountered with k={k}: {e}")
```
